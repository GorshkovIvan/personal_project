```python
class etherscanParser:
    
    def __init__(self, url):
        self.url = url
        self.req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        self.webpage = urlopen(self.req).read()
        self.page_soup = soup(self.webpage, "html.parser")
        
    def get_address_name(self):
        
        title_text = self.page_soup.title.get_text()
        name = ""
        count = 0 
        
        while(not title_text[count].isalnum()):
            count += 1
        while(count < len(title_text) and title_text[count] != "|"):
            if(title_text[count].isalnum() or title_text[count] == " "):
                name += title_text[count]
            count += 1
        self.address_name = name[:len(name)-1]
        
        return self.address_name
    
    def get_labels_through_href(self):
        label = ""
        for a in self.page_soup.find_all('a', href=True):
            if("/accounts/label/" in a['href']):
                label += a['href'][16:]
                label += " "
        
        return label
    
    def get_address_webpage_string(self):
        
        return self.page_soup.prettify()
    
    def get_address_webpage(self):
        
        return self.page_soup
        
```
