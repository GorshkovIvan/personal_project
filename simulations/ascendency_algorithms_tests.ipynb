```python
import import_ipynb
from mixer_ascendency import*
from louvain_ascendency import*
```

```python
mix = 1
step = 0.5
number_of_users = 100
number_of_steps = 1
simulated_networks = []

while(mix <= 1.0):
    world = simulation(number_of_users, 1, 1, mix, number_of_steps)
    world.simulate_random_exchange()
    simulated_networks.append(world.blockchain.network)

    mix += step
```

```python
partitions = []
ascendencies = []
for network in simulated_networks:
    G = louvain_ascendency(network)
    result = G.optimise()
    print(result[0])
    partitions.append(result[0])
    ascendencies.append(result[1])

print(partitions)
print(ascendencies)
```
